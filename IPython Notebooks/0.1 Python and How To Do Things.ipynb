#### Plot example

The code below creates three plots stacked vertically. The purpose here is to demonstrate some features of plotting available through the Python math plotting library, abbreviated 'matplotlib'. Simply highlight the cell below with your left moust button and hold down Shift and hit Enter. The plots should appear below the cell.

```python
# make plots exist inline (not pop-out) and get the numpy and pyplot libraries in place. And make pi easier to write.
%matplotlib inline

import matplotlib.pyplot as plt
import numpy as np

# A double-axis figure (note subplots arg is 2) where everything is symmetric between except 
#   that the first axis ax[0] has a patch and a line thrown in. This shows how matplotlib
#   autoscales the axes to accommodate 'what's there'.

# use myPlotSize value to make plot bigger or smaller
myPlotSize = 11

# fig will be the overall figure
# ax will be a list of 3 subplots ("axes") which are the three plots we see
#   how do we know 3? Because that is the first argument to subplots()
fig, ax = plt.subplots(3, figsize=(myPlotSize, myPlotSize))

x = np.arange(0, 2*np.pi, .1)

# ax[0] is the upper plot; it will have a sine wave
ax[0].plot(x, np.sin(x))

# ax[1] is the lower plot; it will have a cosine wave
ax[1].plot(x, np.cos(x))

# ax[2] is the same as ax[1] for comparison; but using a scatter plot
ax[2].scatter(x, np.cos(x), color='k')

# len(ax) will return 2 because this list has two elements

# a patch is added to plot zero (ax[0]) which will be a circle
ax[0].add_patch(plt.Circle((0, 0), 1, color = 'g', fill=False))

# make the upper plot have equal x~y scale so that the circle looks round
ax[0].axis('equal')

# do this for the third plot ax[2] to compare it with ax[1]
ax[2].axis('equal')

# these are floating-point (float) variables for two points (2.0, 0.4) and (3.5, 1.2)
x1 = 2.0
x2 = 3.5
y1 = 0.4
y2 = 1.1

# add a little line to the upper plot
# this shows how x[] and y[] lists create line plots. 
# scatter plotting uses the same list constructs and plt.scatter()
ax[0].plot([x1, x2], [y1, y2], color='r')

# these lines stretch the x/y ranges to avoid cropping the top subplot
# the result "sort of" listens
ax[0].set_xlim(-2, 7)
ax[0].set_ylim(-2, 2)

# labels for the top subplot
ax[0].set(xlabel='x axis', ylabel = 'y axis', title='some little curves')

# After some playing around: This next line creates space between the subplots so that they
#   can have individual titles that don't overlap the axis label above.
fig.subplots_adjust(left=None, bottom=0.1, right=None, top=0.9, wspace=None, hspace=.4)


# Uncomment these if you would like to see various modifications to the axis range in the upper figure
# ax[0].set_xlim(-7, 7)
# ax[0].set(xlim=(-2, 2), ylim=(-3,3))

# Notice that ax is a List of axes. Here we touch on ax[1]
ax[1].set(xlabel='x', ylabel = 'y')
ax[1].set(title='another curve')

# Notice that ax is a List of axes. Here we touch on ax[1]
ax[2].set(xlabel='the road to Barstow', ylabel = 'road to the coast')
ax[2].set(title='and yet another')
```
