```python
# Kilroy did not write this code (but modified slightly); see pathology remark above
def taylor_compare(func, x0=None, orders=(2, 4), xrange=(0,1), yrange=None, doGrid=true, doLegend=true, doShow=false, npts=200):
    """Compare Taylor series expansion (various orders) to a function.
    Args:
    func : a sympy function
    x0 : float: Ordinate of the Taylor series expansion.  If not given, x0=xrange[0].
    orders : list: List of (integer) orders of Taylor series to show.  Default is (2, 4).
    xrange : 2-tuple or array: Either a (xmin, xmax) tuple (default is (0, 1)) or the actual array
    yrange : 2-tuple(ymin, ymax) tuple indicating the y range; default is the full range of values.
    doGrid: bool (true): Overlay a grid of lines on the plot
    doLegend: bool (true): Superimpose a legend box on the plot (gets messy for large orders!)
    npts : int: Number of points to sample the x range with.  Default is 200.
    """
    if not callable(func):
        raise ValueError('func must be callable')
    if isinstance(xrange, (list, tuple)):
        x = np.linspace(float(xrange[0]), float(xrange[1]), npts)
    else:
        x = xrange
    if x0 is None: x0 = x[0]
        
    # Kilroy: plt.ioff()???
    
    xs = sym.Symbol('x')
    # Make a numpy-callable form of the original function for plotting
    fx = func(xs)
    f = sym.lambdify(xs, fx, modules=['numpy'])
    # We could use latex(fx) instead of str(), but matploblib gets confused
    # with some of the (valid) latex constructs sympy emits.  So we play it safe.
    plt.plot(x, f(x), label=str(fx), lw=2)
    # Build the Taylor approximations, plotting as we go
    apps = {}
    for order in orders:
        app = fx.series(xs, x0, n=order).removeO()
        apps[order] = app
        # Must be careful here: if the approximation is a constant, we can't
        # blindly use lambdify as it won't do the right thing.  In that case, 
        # evaluate the number as a float and fill the y array with that value.
        if isinstance(app, sym.numbers.Number):
            y = np.zeros_like(x)
            y.fill(app.evalf())
        else:
            fa = sym.lambdify(xs, app, modules=['numpy'])
            y = fa(x)
        tex = sym.latex(app).replace('$', '')
        plt.plot(x, y, label=r'$n=%s:\, %s$' % (order, tex) )
        
    # Plot refinements
    if yrange is not None:
        plt.ylim(*yrange)
    if doGrid:
        plt.grid()
    if doLegend:
        plt.legend(loc='best').get_frame().set_alpha(0.8)
    return plt.figure()
```
